### MCMC base functions (likelihood, priors, model)

```python
# easier constants for shorter lines
h = const.h
c = const.c
e = np.e
kb = const.k
```

```python
def model(microns, t_eff, log_factor):
    # compute log flux using Planck's law
    # Planck's law B(lambda, T) = ((2hc^2)/(lambda^5))(1/(exp((hc)/(lambda kb T))-1))
    # We convert into SI for this calculation because scipy constants are in SI
    lambdas = microns / 1e6 #conversion into meters
    flux = ((2. * h * c ** 2) / (lambdas ** 5)) / (e ** ((h * c) / (lambdas * kb * t_eff)) - 1)
    log_flux = np.log10(flux)
    log_flux = log_flux + log_factor
    return log_flux


def log_like(lambda_um, log_flux, log_flux_error, theta, full=False):
    # Compute the log likelihood function
    t_eff, log_factor = theta # unpacking for the model inputs

    if full:
        # from class slide
        first_term = len(lambda_um) * (.5) * np.log10(2 * np.pi) # const
        second_term = np.sum(np.log10(log_flux_error)) # weights
        third_term = (.5) * np.sum((log_flux - model(lambda_um, t_eff, log_factor)) ** 2 / log_flux_error ** 2) #chi sq

        log_likelihood = - first_term - second_term - third_term
        return log_likelihood
    else:
        chi_sq = np.sum(((log_flux - model(lambda_um, t_eff, log_factor)) / log_flux_error) ** 2)
        return chi_sq



def log_prior(theta, theta_shape):
    # calculate the priors
    # theta >> (t_eff, log_factor)
    # theta_shape >> [(t_min, t_max, shape), (log_factor_min, log_factor_max, shape)]
    
    log_priors = []
    t_eff, log_factor = theta

    # temperature
    t_min, t_max, shape = theta_shape[0]
    if shape == 'uniform':
        if t_eff < t_max and t_eff > t_min:
            log_priors.append(0) # because log, this is equal to * 1
        else:
            log_priors.append(1e10) # huge penalty for not being in min < < max
    elif shape == 'loguniform':
        pass
        

    # log_factor
    log_factor_min, log_factor_max, shape = theta_shape[1]
    if shape == 'uniform':
        if log_factor < log_factor_max and log_factor > log_factor_min:
            log_priors.append(0) # because log, this is equal to * 1
        else:
            log_priors.append(1e10) # huge penalty for not being in min < < max
        
    return np.sum(log_priors)
```

### MCMC metropolis rule

```python
def metropolis_rule(likelihood_chain, new_likelihood):
    # metropolis rule for accepting jumps
    # returns the decision and the chain
    current_likelihood = likelihood_chain[-1]

    if new_likelihood < current_likelihood: # accept the jump
        return True
    else:
        jump_prob = np.e ** (-.5 * (new_likelihood - current_likelihood))
        if jump_prob > np.random.uniform(0,1): # accept the jump
            return True
        else: # don't accept
            return False
```

### MCMC main function

```python
def run_mcmc(total_jumps, theta_initial, theta_shape):
    # main function for running MCMC

    # initiallize jump chains
    theta_chain = np.array([np.array(theta_initial)])
    likelihood_chain = np.array([log_like(lambda_um, log_flux, log_flux_err, theta_initial)])
    posterior_chain = np.array([likelihood_chain[0] + log_prior(theta_initial, theta_shape)])

    # jump counter
    j = 0
    while True:
        # get current theta state
        last_t, last_log_factor = theta_chain[-1]
        # obtain theta proposal
        proposal_t = np.random.normal(last_t, t_jump_size)
        proposal_log_factor = np.random.normal(last_log_factor, log_factor_jump_size)
        proposal_theta = (proposal_t, proposal_log_factor)

        # obtain likelihood, prior & posterior for theta proposal
        proposal_likelihood = log_like(lambda_um, log_flux, log_flux_err, proposal_theta)
        proposal_prior = log_prior(proposal_theta, theta_shape)
        proposal_posterior = proposal_likelihood + proposal_prior

        # apply metropolis rule to theta proposal
        decision = metropolis_rule(posterior_chain, proposal_posterior)

        if decision: # we've accepted the jump
            # append proposed state to chain and advancing counter
            theta_chain = np.vstack((theta_chain, proposal_theta))
            likelihood_chain = np.append(likelihood_chain, proposal_likelihood)
            posterior_chain = np.append(posterior_chain, proposal_posterior)
            j = j + 1

            if j%1000 == 0:
                print('Jump number {0}, posterior {1}'.format(j, posterior_chain[-1]))

        if j >= total_jumps:
            break
    
    return theta_chain, likelihood_chain, posterior_chain
```

### save & load mcmc chains

```python
def save_mcmc_chains(theta, likelihood, posterior, chain_name):
    #pickle the mcmc chains so we don't have to run it everytime we restart
    save_path = '../results/chains_{0}.mcmc_chains'.format(chain_name)
    chains = (theta, likelihood, posterior)
    pickle.dump(chains, open(save_path, 'wb'))


def load_mcmc_chains(chain_name):
    #un-pickle the mcmc chains
    load_path = '../results/chains_{0}.mcmc_chains'.format(chain_name)
    theta, likelihood, posterior = pickle.load(open(load_path, 'rb'))
    return theta, likelihood, posterior
```

### base MCMC initialization

```python
# define chain size
total_jumps_base = 10000

# define boundaries for the parameters
t_min, t_max = 150, 550 # taken from lab2.pdf
log_factor_min, log_factor_max = -40, -20 # taken from lab2.pdf

# defind the prior shapes for the parameters
t_shape = 'uniform'
log_factor_shape = 'uniform'
theta_shape_base = [(t_min, t_max, t_shape), (log_factor_min, log_factor_max, log_factor_shape)]

# get initial random theta guess
theta_initial_base = (np.random.uniform(t_min, t_max), np.random.uniform(log_factor_min, log_factor_max))

print('MCMC initial theta: {}'.format(theta_initial_base))
```

### base MCMC proposal jump size

```python
# select jump sizes as fraction of the parameter domains
t_jump_size = .005 * (t_max - t_min)
log_factor_jump_size = .05 * (log_factor_max - log_factor_min)

print('Sigma for jumps are: {0} for t_eff and {1} for log_factor'.format(t_jump_size, log_factor_jump_size))
```

### base MCMC

```python
# base mcmc run
theta_chain_base, likelihood_chain_base, posterior_chain_base = run_mcmc(total_jumps_base, theta_initial_base, theta_shape_base)
```

### save/load base MCMC

```python
# saving results for base mcmc
save_mcmc_chains(theta_chain_base, likelihood_chain_base, posterior_chain_base, 'base')
```

```python
# loading results for base mcmc 
theta_chain_base, likelihood_chain_base, posterior_chain_base = load_mcmc_chains('base')
```

### Multi-seed MCMC

```python
### Multi-seed MCMC
total_jumps = 10000
theta_shape = theta_shape_base

for i in range(10):
    theta_initial = (np.random.uniform(t_min, t_max), np.random.uniform(log_factor_min, log_factor_max))
    theta_chain, likelihood_chain, posterior_chain = run_mcmc(total_jumps, theta_initial, theta_shape)
    save_mcmc_chains(theta_chain, likelihood_chain, posterior_chain, str(i))
```

```python
# load all chains
theta_chain_all = []
likelihood_chain_all = []
posterior_chain_all = []

for i in range(10):
    load_theta_chain, load_likelihood_chain, load_posterior_chain = load_mcmc_chains(str(i))
    theta_chain_all.append(load_theta_chain)
    likelihood_chain_all.append(load_likelihood_chain)
    posterior_chain_all.append(load_posterior_chain)
```

```python
# parameter space walk all chains
fig8 = plt.figure(figsize=(10,5))
ax1 = fig8.add_subplot(111)
for i in range(10):
    theta_chain = theta_chain_all[i]
    ax1.scatter(theta_chain[:,0], theta_chain[:,1], 
                c=np.arange(len(theta_chain)), cmap='coolwarm', alpha=.1)
    ax1.text(theta_chain[0,0], theta_chain[0,1], 'Seed {0}'.format(i))
ax1.set_xlabel('Theta_0: T_eff')
ax1.set_ylabel('Theta_1: Log_factor')
ax1.set_title('MCMC Multi-seed Parameter Space Walk')
fig8.show()
fig8.savefig('../figs/multi_seed_param_walk.pdf')
```
